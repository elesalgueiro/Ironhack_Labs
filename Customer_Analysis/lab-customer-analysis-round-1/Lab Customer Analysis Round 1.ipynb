### 8.Understand the meaning of all columns



_customer_ : this column identifies individual customer (through car plate numbers).

_state_ :in which state the customer has the policy registered. 

_gender_ :- this column identifies the gender of the customer.

_education_ : this column stores the level of education of the customer.

_customer lifetime value_ : total amount of money customer is expected to spend in this company while being a client.

_income_ : income level of the customer.

_monthly premium auto_ : amount of monthly premiums on their policy. 

_number of open complaints_ : this column stores the number of open complaints stored from the customer. 

_policy type_ : type of policy chosen by the customer. It reflects the policy options offered by this company.

_vehicle class_ :type of vehicle (probably the policy types are dependant on this)

_total claim amount_ :total amount claimed by the customer.
