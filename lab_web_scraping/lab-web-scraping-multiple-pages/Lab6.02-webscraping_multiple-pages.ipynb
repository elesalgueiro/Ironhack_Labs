**4.Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe: url = 'https://www.emsc-csem.org/Earthquake/'**

```python
url='https://www.emsc-csem.org/Earthquake/'
response=requests.get(url)
soup= BeautifulSoup(response.content, 'html.parser')    
```

```python
soup.select('td.tabev6 > b > a')[0]
```

```python
#date and time
soup.select('td.tabev6 > b > a')[0].get_text()[-10:-2]
```

```python
#soup.find_all(attrs ={'class':'tabev1'}) --- con find_all
```

```python
#latitude + long
soup.select('td.tabev1')[0:20]
```

```python
#latitude
soup.select("td:nth-child(5)")
```

```python
#longitude
soup.select("td:nth-child(7)")
```

```python
#region
soup.select('td.tb_region')[0].get_text()
```

```python
latitude=[]
longitude=[]
region name=[]
```

```python
date=[]
time=[]
lat=[]
long=[]
region=[]
for i in range(len(soup.select('td.tb_region'))):
    date.append(soup.select('td.tabev6 > b > a')[i].get_text()[0:10])
    time.append(soup.select('td.tabev6 > b > a')[i].get_text()[-10:-2])
    lat.append(soup.select('td.tabev1')[i*2].get_text().replace('\xa0',''))
    long.append(soup.select('td.tabev1')[i*2+1].get_text().replace('\xa0',''))
    region.append(soup.select('td.tb_region')[i].get_text())
```

```python
earthquake=pd.DataFrame({'date':date,
                        'time':time,
                        'region':region,
                        'long':long,
                        'lat':lat})
```

**7.Display the top 10 languages by number of native speakers stored in a pandas dataframe: url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'**

```python
url='https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
response=requests.get(url)
soup= BeautifulSoup(response.content, 'html.parser')  
```

```python
soup.select('tr td:nth-child(2) > a')[0:21][0].get_text()
```

```python
#mw-content-text > div.mw-parser-output > table:nth-child(15) > tbody > tr:nth-child(1) > td:nth-child(3)
```

```python
soup.select('tr> td:nth-child(3)')[0].get_text()
```

```python
languages=[]
speakers=[]
for i in range(20):
    languages.append(soup.select('tr td:nth-child(2) > a')[0:21][i].get_text())
    speakers.append(soup.select('tr> td:nth-child(3)')[0:21][i].get_text().replace('\n', ''))
```

```python
len(speakers)
```

```python
top_10 = pd.DataFrame({'language':languages,
                      'num_speakers':speakers})
```

```python
top_10.head(21)
```
